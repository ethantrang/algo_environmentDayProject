Scraping URLs from The Huffington Post

```python
# time module contains the sleep function
import time
# json module contains functions for us to store our scraped urls in json files
import json
# Beautiful Soup module collects the HTML code from website
from bs4 import BeautifulSoup as bs
# Selenium allows us to execute cross-browser tests
from selenium import webdriver
# Webdriver is an intermediatary between the operating system and the device
from webdriver_manager.chrome import ChromeDriverManager
```

```python
MAIN_URL = "https://www.huffpost.com/" # url of huffington post
urls_list = [] # this list contains the urls being scraped 
```

```python
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.maximize_window()
time.sleep(1)
```

```python
checkpoint = 2000
start = 20000

for i in range (2000,6788):
    
    browser.get(f"https://www.huffpost.com/impact/green?page={i}") # opens previous articles
    get_page = bs(browser.page_source) # collects HTML code

    ### scraps URL ###
    for z in range(0,3): # the website has three content zones, we collect URLS from these zones

        for stories in get_page('div',{'class':'zone__content'})[z]('div',{'class':'card card--standard js-card'}):
            urls_list.append(stories('a')[0]['href']) # collects URL from HTML code

            if len(urls_list) % checkpoint == 0: 

                with open(f"huffpost_article_start={start}_end={start + checkpoint}.json","w") as file:
                    json.dump(urls_list, file, ensure_ascii=True)
                    file.close()

                start += checkpoint
                print(f"--- Checkpoint {start} URLs---") # checkpoint checker

                urls_list.clear()

                time.sleep(2)

    print(f"--- Finished page {i} ---") # page checker

    time.sleep(1)
```
